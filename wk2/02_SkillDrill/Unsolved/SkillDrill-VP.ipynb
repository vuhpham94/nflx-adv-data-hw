```python
# Import Dependencies
import pandas as pd
```

```python
# Get the file path for running team one
filepath = '../Resources/team-1.csv'
```

```python
# Read the CSV file into a DataFrame
team1_df = pd.read_csv(filepath)
team1_df
```

```python
# Melt the DataFrame
team1_df.melt()
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
team1_df.melt(id_vars=['runner_name'])
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
team1_df.melt(id_vars=['coach_name'])
```

```python
# Renaming the columns
team1_df.melt(id_vars=['runner_name'], var_name='runner_info', value_name='runner_value')
```

```python
# Remove coach name from DataFrame
team1_df.drop(columns=['coach_name'])
```

```python
# Show final result
melted_team1_df = team1_df.melt(id_vars=['runner_name'], var_name='runner_info', value_name='runner_value')
melted_team1_df
```

```python
# Save results to a file
melted_team1_df.to_csv('melted_team-1.csv')
```

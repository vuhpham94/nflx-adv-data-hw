```python
# Import dependencies
import pandas as pd
```

```python
filepath = '../Resources/ratings_and_sentiments.csv'
# Read file into DataFrame
ilovecoffee_df = pd.read_csv(filepath)
ilovecoffee_df.head()
```

```python
# Find all reviews with a number rating greater than or equal to 4.
ilovecoffee_df[ilovecoffee_df['num_rating'] >= 4].info()
```

```python
# Find all the coffee shops with a overall sentiment greater than or equal to 4.
ilovecoffee_df[ilovecoffee_df['overall_sent'] >= 4]
```

```python
# Return the count of the above query.
best_coffee_shop = ilovecoffee_df[(ilovecoffee_df['num_rating'] >= 4) & (ilovecoffee_df['overall_sent'] >= 4)]
print("There are {} coffee shops which are the best.".format(len(best_coffee_shop)))
```

```python
# Change the name of the review_text column to review
ilovecoffee_df = ilovecoffee_df.rename(columns={"review_text": "review"})
ilovecoffee_df.head()
```

```python
# Change the name of the bool_HIGH column to bool_high
ilovecoffee_df = ilovecoffee_df.rename(columns={"bool_HIGH": "bool_high"})
ilovecoffee_df.head()
```

```python
# Save your changes to the file as a pipe delimited file.
ilovecoffee_df.to_csv('yelp_rating_with_pipe.csv', sep='|')
```
